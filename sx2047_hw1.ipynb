```python
#3. triangle using law of cosines
side1 = int(input("Enter a side length: "));
side2 = int(input("Enter a second side length: "));
angle = int(input("Enter an angle: "));

import turtle;
import math;
turtle.shape("turtle");
turtle.speed(5);

turtle.forward(side1);
turtle.left(180-angle);
turtle.forward(side2);

#functions work in radians
side3 = math.sqrt(side1**2 + side2**2 - 2*side1*side2*math.cos(angle*math.pi/180));
angle2 = math.acos((side3**2 + side2**2 - side1**2)/(2*side3*side2))*180/math.pi;

turtle.left(180-angle2);
turtle.forward(side3);
```

```python
#5. total time
dayJ = int(input("Please enter the number of days John has worked: "));
hourJ = int(input("Please enter the number of hours John has worked: "));
minJ = int(input("Please enter the number of minutes John has worked: "));

dayB = int(input("Please enter the number of days Bill has worked: "));
hourB = int(input("Please enter the number of hours Bill has worked: "));
minB = int(input("Please enter the number of minutes Bill has worked: "));

totalMin = ((dayJ + dayB)*24 + hourJ + hourB)*60 + minJ + minB;
minTot = totalMin%60;
hourTot = totalMin%(24*60)//60;
dayTot = totalMin//(24*60);

print("The total time both of them worked together is:",dayTot,"days,",hourTot,"hours and",minTot,"minutes.");
```
